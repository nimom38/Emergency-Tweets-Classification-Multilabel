```python
for i in range(15):
  file_name = ss[i]

  dat=list()
  with open (file_name, 'r') as f:
      for l in f.readlines():
          if not l.strip (): # skip empty lines
              continue

          json_data = json.loads (l)
          dat.append(json_data)

  for j in range(len(dat)):
    dat[j]['identifier'] = dat[j]['allProperties']['id']
  tweet_data_frame = pd.DataFrame.from_dict(dat)
  tweet_data_frame.set_index('identifier')

  tweet_data_frame.to_csv(tt[i])
  files.download(tt[i])
```
