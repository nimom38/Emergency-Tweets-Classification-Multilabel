```python
def func2(a, b):
  for index, value in a.items():
    huhu = value
    if str(int(huhu)) == str(int(b)):
      return index
```

```python
def check(a):
  a = str(a)
  for ch in a:
    if ch >= '0' and ch <= '9':
      continue
    if ch == '.':
      continue
    return False
  return True
```

```python
import json
import math
import numpy
cc = 0
combo = pd.read_csv('Combined labels.csv')
print(combo.columns)
hoho = combo['postID']
store = []
file_names = [
              'trecis2019-A-test.shootingDallas2017.csv',
              'trecis2019-A-test.hurricaneFlorence2018.csv',
              'trecis2019-A-test.floodChoco2019.csv',
              'trecis2019-A-test.fireYMM2016.csv',
              ##'trecis2019-A-test.fireAndover2019.csv',
              'trecis2019-A-test.earthquakeCalifornia2014.csv',
              'trecis2019-A-test.earthquakeBohol2013.csv'
]
assert len(file_names) == 6
```

```python
store.clear()
for file in file_names:
  cc += 1
  df = pd.read_csv(file)
  arr = []
  for row in df.itertuples(name=None):
    if (row[6] is None) or ((not isinstance(row[6], str)) and (math.isnan(row[6]) == True)):
      continue
    if (row[6] == "TWEETS"):
      continue
    if check(row[6]) == False:
      continue
    if func(hoho, row[6]) == False:
      continue
    yap = func2(hoho, row[6])
    xxx = None
    if (row[2] is not None) and (isinstance(row[2], str)):
      xxx = eval(row[2])
    dic = {}
    dic['tweet_text'] = None
    if xxx is not None:
      assert 'text' in xxx
      dic['tweet_text'] = xxx['text']
    dic['user_verified'] = None
    if xxx is not None:
      if 'user.verified' in xxx:
        dic['user_verified'] = int(xxx['user.verified']=='true')
    dic['user_followers'] = None
    if xxx is not None:
      if 'user.followersCount' in xxx:
        dic['user_followers'] = int(xxx['user.followersCount'])
      if 'user.followers_count' in xxx:
        dic['user_followers'] = int(xxx['user.followers_count'])
    dic['user_following'] = None
    dic['media'] = 0
    if xxx is not None:
      if 'mediaEntities' in xxx:
        dic['media'] = int(len(xxx['mediaEntities'])>2)
      if 'entities.media' in xxx:
        dic['media'] = int(len(xxx['entities.media'])>2)
      if 'retweeted_status.extended_entities.media' in xxx:
        dic['media'] = int(len(xxx['retweeted_status.extended_entities.media'])>2)
    dic['user_statuses'] = None
    if xxx is not None:
      if 'user.statusesCount' in xxx:
        dic['user_statuses'] = int(xxx['user.statusesCount'])
      if 'user.statuses_count' in xxx:
        dic['user_statuses'] = int(xxx['user.statuses_count'])
    dic['timestamp'] = None
    if xxx is not None:
      if 'timestamp' in xxx:
        dic['timestamp'] = xxx['timestamp']
    dic['identifier'] = str(row[6])
    dic['language'] = None
    if xxx is not None:
      if 'user.lang' in xxx:
        dic['language'] = xxx['user.lang']
    dic['location_metadata'] = None
    if xxx is not None:
      if 'place.boundingBoxCoordinates' in xxx:
        dic['location_metadata'] = xxx['place.boundingBoxCoordinates']
    for words in clss:
      dic[words] = 0
    hmm = int(row[6])
    temp = []
    temp = combo.iloc[yap, 5]
    temp2 = ""
    for i in range(len(temp)):
      if (temp[i] == '[') or (temp[i] == ']') or (temp[i] == ',') or (temp[i] == '\''):
        temp2 += ' '
      else:
        temp2 += temp[i]
    temp2 = list(temp2.split())
    for words in temp2:
      dic[words] = 1
    dic['priority_level'] = combo.iloc[yap, 6]
    arr.append(dic)
  df = pd.DataFrame.from_dict(arr)
  df.drop_duplicates(subset=['identifier'], keep = "first")
  df.set_index('identifier', inplace=True, drop=True)
  store.append(df)
  print(cc, file)
  print(df.shape)
  print(df.columns)
  print(df.info())
  #df.to_csv(file)
  #files.download(file)
```
