```python
def func(a, b):
  for index, value in a.items():
    huhu = value
    if str(int(huhu)) == str(int(b)):
      return True
  return False
```

```python
def func2(a, b):
  for index, value in a.items():
    huhu = value
    if str(int(huhu)) == str(int(b)):
      return index
```

```python
import json
import math
combo = pd.read_csv('Combined labels.csv')
print(combo.columns)
hoho = combo['postID']
file_names = [
              'westTexasExplosion2013.csv',
              'typhoonPablo2012.csv',
              'laAirportShooting2013.csv',
              'floodColorado2013.csv',
              'fireColorado2012.csv',
              'costaRicaEarthquake2012.csv'
]
for file in file_names:
  df = pd.read_csv(file)
  arr = []
  for row in df.itertuples(name=None):
    if (row[2] is None) or ((not isinstance(row[2], str)) and (math.isnan(row[2]) == True)):
      continue
    if (row[2] == "TWEETS"):
      continue
    if func(hoho, row[2]) == False:
      continue
    yap = func2(hoho, row[2])
    xxx = None
    yyy = None
    zzz = None
    if (row[6] is not None) and (isinstance(row[6], str)):
      xxx = eval(row[6])
    if (row[10] is not None) and (isinstance(row[10], str)):
      yyy = eval(row[10])
    if (row[9] is not None) and (isinstance(row[9], str)):
      zzz = eval(row[9])
    dic = {}
    dic['tweet_text'] = row[3]
    dic['user_verified'] = None
    if xxx is not None:
      dic['user_verified'] = int(xxx['user.verified']=='true')
    dic['user_followers'] = None
    if yyy is not None:
      dic['user_followers'] = int(yyy['User Followers'])
    dic['user_following'] = None
    if yyy is not None:
      dic['user_following'] = int(yyy['User Follows'])
    dic['media'] = 0
    if (row[11] is not None) and (isinstance(row[11], str)):
      dic['media'] = int(len(row[11])>2)
    dic['user_statuses'] = None
    if yyy is not None:
      dic['user_statuses'] = int(yyy['User Statuses'])
    dic['timestamp'] = None
    if (row[5] is not None) and (isinstance(row[5], str) or (not isinstance(row[5], str) and not math.isnan(row[5]))):
      dic['timestamp'] = str(row[5])
    dic['identifier'] = str(row[2])
    dic['language'] = None
    if (row[7] is not None) and (isinstance(row[7], str)):
      dic['language'] = row[7]
    dic['location_metadata'] = None
    if zzz is not None:
      dic['location_metadata'] = zzz['geolocation']
    for words in clss:
      dic[words] = 0
    hmm = int(row[2])
    temp = []
    temp = combo.iloc[yap, 5]
    temp2 = ""
    for i in range(len(temp)):
      if (temp[i] == '[') or (temp[i] == ']') or (temp[i] == ',') or (temp[i] == '\''):
        temp2 += ' '
      else:
        temp2 += temp[i]
    temp2 = list(temp2.split())
    for words in temp2:
      dic[words] = 1
    dic['priority_level'] = combo.iloc[yap, 6]
    arr.append(dic)
  df = pd.DataFrame.from_dict(arr)
  df.set_index('identifier', inplace=True, drop=True)
  df.to_csv(file)
  files.download(file)
```
