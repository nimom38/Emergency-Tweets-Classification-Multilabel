```python
import pandas as pd
import json
```

```python
ss = [
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidDC2020.json",
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidNYC2020.json",
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidWashingtonState2020.json"
]
tt = [
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidDC2020.csv",
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidNYC2020.csv",
  "/content/drive/MyDrive/Colab-Notebooks/preprocess/CSV/7. TestSet_2020_A_Covid/trecis2020-A-covid.covidWashingtonState2020.csv"
]
```

```python
for i in range(3):
  file_name = ss[i]

  dat=list()
  cc = 0
  with open (file_name, 'r') as f:
      for l in f.readlines():
          if not l.strip (): # skip empty lines
              continue

          json_data = json.loads (l)
          dat.append(json_data)
          cc += 1
          if cc > 500:
            break

  for j in range(len(dat)):
    dat[j]['identifier'] = dat[j]['allProperties']['id']
  tweet_data_frame = pd.DataFrame.from_dict(dat)
  tweet_data_frame.set_index('identifier')

  tweet_data_frame.to_csv(tt[i])
  files.download(tt[i])
```
