```python
import pandas as pd
import json
```

```python
file_name = "trecis2019-B-test.cycloneKenneth2019.json"

dat=list()
with open (file_name, 'r') as f:
    for l in f.readlines():
        if not l.strip (): # skip empty lines
            continue

        json_data = json.loads (l)
        dat.append(json_data)

dat
```

```python
for i in range(len(dat)):
  dat[i]['identifier'] = dat[i]['allProperties']['id']
tweet_data_frame = pd.DataFrame.from_dict(dat)
tweet_data_frame.set_index('identifier')
```

```python
from google.colab import files
tweet_data_frame.to_csv('trecis2019-B-test.cycloneKenneth2019.csv')
```
